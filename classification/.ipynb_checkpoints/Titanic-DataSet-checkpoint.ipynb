# preprocessing

```python
x = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
x_train = pd.get_dummies(x_train,columns=["Sex","Embarked"]) 
x_test = pd.get_dummies(x_test,columns=["Sex","Embarked"])
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

# model training

```python
model = LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
p_test = model.predict(x_test)
p_train = model.predict(x_train)

#scaled_model = model.fit(x_train_scaled,y_train)
#scaled_test_p = model.predict(x_test_scaled)
#scaled_train_p  = model.predict(x_train_scaled)
```

# evaluation

```python
accuracy_test = accuracy_score(y_test,p_test)
accuracy_train = accuracy_score(y_train,p_train)
#accuracy_test_scaled = accuracy_score(y_test,scaled_test_p)
#accuracy_train_scaled = accuracy_score(y_train,scaled_train_p)
cm_test = confusion_matrix(y_test,p_test)
cm_train = confusion_matrix(y_train,p_train)
#cm_scaled_test = confusion_matrix(y_test,scaled_test_p)
#cm_scaled_train = confusion_matrix(y_train,scaled_train_p)
```

```python
print("accuracy test:",accuracy_test)
print("accuracy train:",accuracy_train)
#print("scaled test accuracy:",accuracy_test_scaled)
#print("scaled train accuracy:",accuracy_train_scaled)
print()
print("cm train:")
print(cm_train)

print("cm test:")
print(cm_test)

#print("cm scaled test:",cm_scaled_test)
#print("cm scaled train:",cm_scaled_train)
```

```python
print("test")
print(classification_report(y_test,p_test))

print("train")
print(classification_report(y_train,p_train))
```

```python
y_prob = model.predict_proba(x_test)[:,1]
fpr,tpr,thresholds = roc_curve(y_test,y_prob)

auc_score = roc_auc_score(y_test,y_prob)
print(auc_score)

plt.plot(fpr,tpr)
plt.plot([0,1],[0,1],linestyle='--')
plt.xlabel("false positive rate")
plt.ylabel("true positive rate")
plt.title("roc curve")
plt.show()
```

# knn model

```python
model2 = KNeighborsClassifier(n_neighbors=5)
model2.fit(x_train_scaled,y_train)

knn_p_train = model2.predict(x_train_scaled)
knn_p_test = model2.predict(x_test_scaled)
```

```python
accuracy_test_knn= accuracy_score(y_test,knn_p_test)
accuracy_train_knn = accuracy_score(y_train,knn_p_train)
#cm_scaled_test = confusion_matrix(y_test,knn_p_test)
#cm_scaled_train = confusion_matrix(y_train,knn_p_train)
```

```python
print("train accuracy")
print(accuracy_train_knn)

print("test accuracy")
print(accuracy_test_knn)
```

# optimal k

```python
train_scores = []
test_scores = []

k_values = range(1,21)
for k in k_values:
    model3 = KNeighborsClassifier(
    n_neighbors=k)
    
    model3.fit(x_train_scaled,y_train)
    train_scores.append(model3.score(x_train_scaled,y_train))
    test_scores.append(model3.score(x_test_scaled,y_test))
    
for k,t in zip(k_values,test_scores):
    print(k,round(t,3))
```

# Support vector machine

```python
svm_model = SVC(kernel='linear',random_state=42)
para_grid = {'C':[0.01,0.1,1,10,100]}
grid = GridSearchCV(estimator = svm_model,
                   param_grid = para_grid,
                   cv = 5,
                   scoring = "accuracy")
grid.fit(x_train_scaled,y_train)

print("best c:",grid.best_params_)
print("best scores:",grid.best_score_)
```

```python
best_model = grid.best_estimator_
svm_pred_test = best_model.predict(x_test_scaled)
svm_pred_train = best_model.predict(x_train_scaled)
```

```python
svm_acc_train = accuracy_score(y_train,svm_pred_train)
svm_acc_test = accuracy_score(y_test,svm_pred_test)

print("train svm accuracy:",svm_acc_train)
print("test svm accuracy:",svm_acc_test)
```

```python
svm_model2 = SVC(kernel='rbf',random_state=42)
para_grid2 = {'C':[0.1,1,10,100],
            'gamma':[0.001,0.01,0.1,1]}
grid2 = GridSearchCV(estimator = svm_model2,
                   param_grid = para_grid2,
                   cv = 5,
                   scoring = "accuracy")
grid2.fit(x_train_scaled,y_train)

print("best c:",grid2.best_params_)
print("best scores:",grid2.best_score_)

```

```python
best_model2 = grid2.best_estimator_
svm_pred_test_rbf = best_model2.predict(x_test_scaled)
svm_pred_train_rbf = best_model2.predict(x_train_scaled)
```

```python
svm_acc_train_rbf = accuracy_score(y_train,svm_pred_train_rbf)
svm_acc_test_rbf = accuracy_score(y_test,svm_pred_test_rbf)

print("train svm accuracy:",svm_acc_train_rbf)
print("test svm accuracy:",svm_acc_test_rbf)
```

# Decision tree for classification 

```python
dt_model = DecisionTreeClassifier(random_state=42)
para_grid = {"criterion":["gini","entropy"],
             "max_depth":[3,5,7,10,15],
            "min_samples_split":[2,5,10],
            "min_samples_leaf":[1,2,4],
            "max_features":[None,"sqrt","log2"]}

dt_grid = GridSearchCV(estimator = dt_model,
                   param_grid = para_grid,
                   cv = 5,
                   scoring = "accuracy")
dt_grid.fit(x_train,y_train)
print("best scores:",grid.best_score_)
```

```python
best_dt_model = dt_grid.best_estimator_
dt_pred_train = best_model.predict(x_train)
dt_pred_test = best_model.predict(x_test)
prob_train = best_dt_model.predict_proba(x_train)
prob_test = best_dt_model.predict_proba(x_test)

#print("probability of train:",prob_train)
print("probability of test:\n",prob_test[:5])
```

```python
dt_acc_train = accuracy_score(y_train,dt_pred_train)
dt_acc_test = accuracy_score(y_test,dt_pred_test)

print("train decision tree accuracy:",dt_acc_train)
print("test decision tree accuracy:",dt_acc_test)
```

# random forest for classification

```python
rf_model = RandomForestClassifier(random_state = 42)
rf_para_grid = {"criterion":["gini"],
                "n_estimators":[100],
             "max_depth":[10],
            "min_samples_split":[5],
            "min_samples_leaf":[2],
            "max_features":["sqrt","log2"],
            "bootstrap":[True],
            "oob_score" :[True]}

rf_grid = GridSearchCV(estimator = rf_model,
                   param_grid = rf_para_grid,
                   cv = 5,
                   scoring = "accuracy",
                   n_jobs=-1)
rf_grid.fit(x_train,y_train)
print("best parameters:",rf_grid.best_params_)
print("best scores:",grid.best_score_)
```

```python
best_rf_model = rf_grid.best_estimator_
best_rf_model.fit(x_train,y_train)
rf_pred_train = best_rf_model.predict(x_train)
rf_pred_test = best_rf_model.predict(x_test)
#prob_train = best_dt_model.predict_proba(x_train)
#prob_test = best_dt_model.predict_proba(x_test)
```

```python
rf_acc_train = accuracy_score(y_train,rf_pred_train)
rf_acc_test = accuracy_score(y_test,rf_pred_test)

print("train decision tree accuracy:",rf_acc_train)
print("test decision tree accuracy:",rf_acc_test)
```
