# Data Cleaning

```python
a = (df.isna().sum()/df.shape[0])*100
a
```

```python
df.nunique()
```

```python
b = df._get_numeric_data()
b.head()
```

```python
b.isna().sum()/b.shape[0]*100
```

```python
b['Lot Frontage'] = b['Lot Frontage'].fillna(b['Lot Frontage'].median())
b['Year Remod/Add'] = b['Year Remod/Add'].fillna(b['Year Remod/Add'].median())
b['Mas Vnr Area'] = b['Mas Vnr Area'].fillna(b['Mas Vnr Area'].median())
b['BsmtFin SF 1'] = b['BsmtFin SF 1'].fillna(b['BsmtFin SF 1'].median())
b['BsmtFin SF 2'] = b['BsmtFin SF 2'].fillna(b['BsmtFin SF 2'].median())
b['Bsmt Unf SF'] = b['Bsmt Unf SF'].fillna(b['Bsmt Unf SF'].median())
b['Total Bsmt SF'] = b['Total Bsmt SF'].fillna(b['Total Bsmt SF'].median())
b['Garage Yr Blt'] = b['Garage Yr Blt'].fillna(b['Garage Yr Blt'].median())     
b['Garage Cars'] = b['Garage Cars'].fillna(b['Garage Cars'].mode()[0])       
b['Garage Area'] = b['Garage Area'].fillna(b['Garage Area'].median())    
b['Bsmt Full Bath'] = b['Bsmt Full Bath'].fillna(b['Bsmt Full Bath'].mode()[0]) 
b['Bsmt Half Bath'] = b['Bsmt Half Bath'].fillna(b['Bsmt Half Bath'].mode()[0])
```

```python
b.isna().sum()
```

```python
b.duplicated().sum()
```

```python
df.describe(include = 'all')
```

```python
c = df.drop(columns=df._get_numeric_data().columns)

c.isna().sum()/c.shape[0]*100
```

```python
d=c.drop(columns=['Pool QC','Fence','Misc Feature','Alley','Fireplace Qu'])
d.isna().sum()/c.shape[0]*100
```

```python
d['Bsmt Qual'] = d['Bsmt Qual'].fillna(d['Bsmt Qual'].mode()[0])
d['Bsmt Cond'] = d['Bsmt Cond'].fillna(d['Bsmt Cond'].mode()[0])
d['Bsmt Exposure'] = d['Bsmt Exposure'].fillna(d['Bsmt Exposure'].mode()[0])
d['BsmtFin Type 1'] = d['BsmtFin Type 1'].fillna(d['BsmtFin Type 1'].mode()[0])
d['BsmtFin Type 2'] = d['BsmtFin Type 2'].fillna(d['BsmtFin Type 2'].mode()[0])
d['Mas Vnr Type'] = d['Mas Vnr Type'].fillna(d['Mas Vnr Type'].mode()[0])
d['Electrical'] = d['Electrical'].fillna(d['Electrical'].mode()[0])
d['Garage Type'] = d['Garage Type'].fillna(d['Garage Type'].mode()[0])
d['Garage Finish'] = d['Garage Finish'].fillna(d['Garage Finish'].mode()[0])     
d['Garage Qual'] = d['Garage Qual'].fillna(d['Garage Qual'].mode()[0])        
d['Garage Cond'] = d['Garage Cond'].fillna(d['Garage Cond'].mode()[0])       
```

```python
d.isna().sum() 
```
